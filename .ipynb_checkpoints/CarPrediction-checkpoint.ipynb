```python
final_dataset=df[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven',
       'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner', 'Build_on']]
```

```python
final_dataset.head()
```

```python
dataset=final_dataset[['Selling_Price', 'Present_Price', 'Kms_Driven',
       'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner', 'Build_on']]
```

```python
dataset.head()
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
cormat=dataset.corr()
top_corr_feature=cormat.index
plt.figure(figsize=(20,20))
#heat map for correlation
g=sns.heatmap(dataset[top_corr_feature].corr(),annot=True,cmap="RdYlGn")
```

```python
#Writing Model
dataset.head()
```

```python
X=dataset.iloc[:,1:]
Y=dataset.iloc[:,0]
```

```python
X.head()
```

```python
Y.head()
```

```python
### Feature Importance
from sklearn.ensemble import ExtraTreesRegressor
```

```python
model=ExtraTreesRegressor()
model.fit(X,Y)
```

```python
print(model.feature_importances_)
```

```python
feat_importances=pd.Series(model.feature_importances_,index=X.columns)
feat_importances.nlargest(5).plot(kind='barh')
plt.show()
```
